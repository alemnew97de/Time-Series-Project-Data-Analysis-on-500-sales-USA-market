```python
import pandas as pd
import zipfile

# Define the path to the ZIP archive
zip_path = '/content/individual_stocks_5yr-20240225T141507Z-001.zip'

# Create an empty DataFrame to store the combined data
full_df = pd.DataFrame()

# Open the ZIP archive
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    # Get the list of files in the archive
    files_in_zip = zip_ref.namelist()

    # Iterate through each file in the archive
    for file_name in files_in_zip:
        # Check if the file is a CSV file
        if file_name.endswith('.csv'):
            # Read the CSV file into a DataFrame
            with zip_ref.open(file_name) as csv_file:
                current_df = pd.read_csv(csv_file, encoding='iso-8859-1', error_bad_lines=False)

                # Concatenate the current DataFrame with the full DataFrame
                full_df = pd.concat([full_df, current_df], ignore_index=True)

# Display the combined DataFrame
print(full_df)
```

## 4.. analyse Closing price change in apple stock !
##### Daily Stock Return Formula
    To calculate how much you gained or lost per day for a stock, subtract the opening price from the closing price. Then, multiply the result by the number of shares you own in the company.

```python
company_list
```

```python
import zipfile
import pandas as pd

# Define the path to the ZIP archive
zip_path = '/content/individual_stocks_5yr-20240225T141507Z-001.zip'

# Define the path to the CSV file within the ZIP archive
csv_file_path = 'individual_stocks_5yr/UPS_data.csv'

# Define a function to read the CSV file from the ZIP archive
def read_csv_from_zip(zip_path, csv_file_path):
    # Open the ZIP archive
    with zipfile.ZipFile(zip_path, 'r') as zip_ref:
        # Check if the CSV file exists in the ZIP archive
        if csv_file_path in zip_ref.namelist():
            # Extract the CSV file from the ZIP archive and read it into a DataFrame
            with zip_ref.open(csv_file_path) as csv_file:
                df = pd.read_csv(csv_file)
            return df
        else:
            print(f"CSV file '{csv_file_path}' not found in the ZIP archive.")

# Call the function to read the CSV file from the ZIP archive
ups_data_df = read_csv_from_zip(zip_path, csv_file_path)

# Display the DataFrame
print(ups_data_df)
```

```python
import zipfile
import pandas as pd

# Define the path to the ZIP archive
zip_path = '/content/individual_stocks_5yr-20240225T141507Z-001.zip'

# Define a list of CSV files to read from the ZIP archive
csv_files = [
    'individual_stocks_5yr/VLO_data.csv',
    'individual_stocks_5yr/MO_data.csv',
    'individual_stocks_5yr/UTX_data.csv',
    'individual_stocks_5yr/XOM_data.csv'
]

# Define a function to read multiple CSV files from the ZIP archive
def read_csv_files_from_zip(zip_path, csv_files):
    dfs = []  # List to store DataFrames

    # Open the ZIP archive
    with zipfile.ZipFile(zip_path, 'r') as zip_ref:
        for csv_file in csv_files:
            # Check if the CSV file exists in the ZIP archive
            if csv_file in zip_ref.namelist():
                # Extract the CSV file from the ZIP archive and read it into a DataFrame
                with zip_ref.open(csv_file) as csv_file_obj:
                    df = pd.read_csv(csv_file_obj)
                    dfs.append(df)
            else:
                print(f"CSV file '{csv_file}' not found in the ZIP archive.")

    return dfs

# Call the function to read multiple CSV files from the ZIP archive
dfs = read_csv_files_from_zip(zip_path, csv_files)

# Display the DataFrames
for df in dfs:
    print(df)
```

```python
(closing_price['apple_close'] - closing_price['apple_close'].shift(1))/closing_price['apple_close'].shift(1) * 100
```

```python
for col in closing_price.columns:
    closing_price[col + '_pct_change'] = (closing_price[col] - closing_price[col].shift(1))/closing_price[col].shift(1) * 100
```

```python
closing_price
```

```python
closing_price.columns
```

```python
clsing_p = closing_price[['apple_close_pct_change', 'amzn_close_pct_change',
       'goog_close_pct_change', 'msft_close_pct_change']]
```

```python
clsing_p
```

```python
### since we have used Pairplot already , lets use extension of Pairplot , ie Pairgrid :
```

```python
'''
Pairplot : we have histogram on diagonals & scatterplot/kde/
            any_other_plot which tells dist.. on rest of the plot


Pairgrid : Once we create grid , we can set plot as per our need :

ie , if we have 4 features , it creates total 16 graphs/plots or matrices of 4*4



    There would be various possibilities for type of plots in our Pairgrid which we can set as per our need :

    a) all plots can be scatterplot
    b) on diagonal , we have histogram & rest will be scatterplot
    c) on diagonal , we have histogram & rest will be kdeplot
    c) on diagonal , we have histogram & below diagonal will be kdeplot
            & upper diagonal will be scatterplot

Note :: kdeplot for 2 features also known as contour plots which returns density
but kdeplot returns distribution if we are performing univariate analysis else it will show density ..

'''
```

```python
g = sns.PairGrid(data= clsing_p)
g.map_diag(sns.histplot)
g.map_lower(sns.scatterplot)
g.map_upper(sns.kdeplot)
```

```python
'''

Conclusion :
While Comparing 'AAPL_close_pct_change' to 'AMZN_close_pct_change'  , it shows a linear relationship upto some extent..


'''
```
